Kolon baz alınarak verilerin %30'u eksik ise o kolonu kaldırır.

```python
tr = len(df) * 0.7
df.dropna(thresh = tr, axis = 1, inplace = True)
df
```

Satırlardaki boş değerler %20 veya daha fazlaysa o satırı kaldırır.

```python
perc = 20.0 
min_count =  int(((100-perc)/100)*df.shape[1] + 1)
df2 = df.dropna( axis=0, 
                    thresh=min_count)

df2
```

Kalan boş değerlerimize baktığımızda;

```python
print(df2.isnull().sum())
```

En çok tekrar eden sayıya göre(mod) verileri doldurmak

```python
	
df2['continent'] = df2['continent'].fillna(df2['continent'].mode()[0])
df2
```

Ortalama değer ile verileri doldurmak.

```python
df2['new_cases'] = df2['new_cases'].fillna(df2['new_cases'].mean())
df2['new_deaths'] = df2['new_deaths'].fillna(df2['new_deaths'].mean())
df2['reproduction_rate'] = df2['reproduction_rate'].fillna(df2['reproduction_rate'].mean())
df2['stringency_index'] = df2['stringency_index'].fillna(df2['stringency_index'].mean())
df2['population_density'] = df2['population_density'].fillna(df2['population_density'].mean())
df2['aged_65_older'] = df2['aged_65_older'].fillna(df2['aged_65_older'].mean())
df2['aged_70_older'] = df2['aged_70_older'].fillna(df2['aged_70_older'].mean())
df2['gdp_per_capita'] = df2['gdp_per_capita'].fillna(df2['gdp_per_capita'].mean())
df2['cardiovasc_death_rate'] = df2['cardiovasc_death_rate'].fillna(df2['cardiovasc_death_rate'].mean())
df2['diabetes_prevalence'] = df2['diabetes_prevalence'].fillna(df2['diabetes_prevalence'].mean())
df2['human_development_index'] = df2['human_development_index'].fillna(df2['human_development_index'].mean())
```
