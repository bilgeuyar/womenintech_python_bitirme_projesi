```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print(tf.isnull().sum())
```

```python
tr = len(tf) * 0.7
tf.dropna(thresh = tr, axis = 1, inplace = True)
tf
```

```python
perc = 20.0 
min_count =  int(((100-perc)/100)*tf.shape[1] + 1)
tf2 = tf.dropna( axis=0, 
                    thresh=min_count)

tf2
```

```python
print(tf2.isnull().sum())
```

```python
tf2.columns
```

***ANALİZLER***

Tarihe Göre Yeni Vaka Sayıları

```python
plt.figure(figsize=(30,6))
plt.xticks(rotation=90)

plt.plot(tf2.date, tf2.new_cases)
plt.title("Tarihe Göre Yeni Vaka Sayıları")
plt.xlabel("Tarih")
plt.ylabel("Vaka Sayıları")

plt.show()
```

Pozitif Çıkan Sonuçlar ve Yeni Vakalar Arasındaki İlişki

```python
sns.scatterplot(x="positive_rate", y="new_cases", data=tf2)
```

Toplam Vaka ve Toplam Ölüm Arasındaki İlişki

```python
sns.lineplot(x='total_cases', y='total_deaths', data=tf2)
```
